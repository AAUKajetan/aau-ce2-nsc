## Simple Monte Carlo Pi Calculation (single process)

```python
# Generated by Claude Sonnet 4.5 on 2026-03-02
import numpy as np
import matplotlib.pyplot as plt

# Simple Monte Carlo Pi Calculation (Single-threaded)
print("=" * 70)
print("Monte Carlo Method: How It Works")
print("=" * 70)

# Generate random points
n_samples = 10000
# np.random.seed(42)  # For reproducibility

# Generate random points in [0,1] x [0,1]
x = np.random.random(n_samples)
y = np.random.random(n_samples)

# Check which points fall inside the quarter circle (x^2 + y^2 <= 1)
distances = x**2 + y**2
inside_circle = distances <= 1

# Count points inside
points_inside = np.sum(inside_circle)
points_total = n_samples

# Calculate Pi estimate
pi_estimate = 4 * points_inside / points_total

print(f"\nGenerated {n_samples:,} random points in unit square [0,1] x [0,1]")
print(f"Points inside quarter circle: {points_inside:,}")
print(f"Points outside quarter circle: {points_total - points_inside:,}")
print(f"\nRatio: {points_inside}/{points_total} = {points_inside/points_total:.6f}")
print(f"\nπ estimate = 4 × {points_inside/points_total:.6f} = {pi_estimate:.6f}")
print(f"True π value = {np.pi:.6f}")
print(f"Error = {abs(pi_estimate - np.pi):.6f}")
print("=" * 70)
```
