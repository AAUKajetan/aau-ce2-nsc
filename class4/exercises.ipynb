In this exercise, we continue working with the Pi calculation script from lecture 4. We will measure execution time for different chunk sizes and analyze the performance impact on both single-core and multi-core systems.

## Tasks

1. **Extend the Pi calculation script** to measure execution time for different chunk sizes:
    - Let L = 10, 100, 1000, ..., 1e6
    - Keep total trials constant: N*L = 1e6, so N = 1e6/L

2. **Single-core analysis (P=1)**:
    - Determine the optimal chunk size
    - Plot results using `plt.semilogx()`

3. **Multi-core analysis (P=8 or your machine's core count)**:
    - Compare optimal chunk size with single-core results
    - Explain the difference between single and multi-core performance

4. **Analysis**:
    - Investigate why chunk size affects performance differently on single vs. multi-core systems
    - Consider how processes are allocated to cores and overhead of process management

# Exercise 5.2: Map-Filter-Reduce Programming

Master the Map-Filter-Reduce paradigm through single-core and multiprocessing implementations.

## Single-core Version

1. **Input**: Generate N random integers in the range [10, 100]
2. **Map-Filter-Reduce workflow**:
    - **Map**: Subtract 7 from all values
    - **Filter**: Keep only odd numbers
    - **Reduce**: Calculate the sum
3. Use Python's built-in `map()`, `filter()`, and `reduce()` functions

## Multiprocessing Version

1. Extend the program to use `pool.map()` for parallel execution
2. Note: This problem is too small to show significant performance benefits
3. Limitation: Multiprocessing natively supports only `map()`, not complete map-filter-reduce workflows
4. Consider: Advanced libraries for full map-filter-reduce parallelization exist but are not in conda
